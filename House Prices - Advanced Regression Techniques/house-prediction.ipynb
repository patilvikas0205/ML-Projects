```python
from sklearn.feature_selection import SelectKBest, chi2
```

```python
fs_model=SelectKBest(chi2,k=110)
train_new=fs_model.fit_transform(train_data,y)
test_new=fs_model.transform(test_data)
```

```python
from xgboost import XGBRegressor
xgb_model=XGBRegressor(n_estimators=500,learning_rate=0.1,max_depth=3,min_child_weight=1,seed=100)
xgb_model.fit(train_data.drop(["Id"],axis=1),y)
xgb_model.score(train_data.drop(["Id"],axis=1),y)
```

```python
xgb_model.score(test_data.drop(["Id"],axis=1),xgb_model.predict(test_data.drop(["Id"],axis=1)))
```

```python
from sklearn.model_selection import RandomizedSearchCV
parameters = {#'nthread':[4], #when use hyperthread, xgboost may become slower
              #'objective':['reg:linear'],
              'learning_rate': [0.03, 0.05, 0.07], #so called `eta` value
              'max_depth': [3,4],
              'min_child_weight': [2],
              #'silent': [1],
              #'subsample': [0.7],
              #'colsample_bytree': [0.7],
              'n_estimators': [250]}

xgb_grid = RandomizedSearchCV(xgb_model,
                        parameters,
                        cv = 10,
                        #n_jobs = 5,
                        verbose=True)

xgb_grid.fit(train_data.drop(["SaleCondition_Alloca","Id"],axis=1),y)
```

```python
xgb_grid.best_params_
```

```python
xgb_grid.score(train_data.drop(["SaleCondition_Alloca","Id"],axis=1),xgb_grid.predict(train_data.drop(["SaleCondition_Alloca","Id"],axis=1)))
```

```python
xgb_grid.score(test_data.drop(["SaleCondition_Alloca","Id"],axis=1),xgb_grid.predict(test_data.drop(["SaleCondition_Alloca","Id"],axis=1)))
```
